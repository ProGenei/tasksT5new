```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

def visualize_traffic_over_time(df, datetime_col, traffic_col):
    # Ensure the datetime column is in datetime format
    df[datetime_col] = pd.to_datetime(df[datetime_col])

    # Sort the dataframe by the datetime column
    df = df.sort_values(by=datetime_col)

    # Set up the plot
    plt.figure(figsize=(15, 8))

    # Create the main time series plot
    sns.lineplot(data=df, x=datetime_col, y=traffic_col)

    # Customize the plot
    plt.title(f'{traffic_col} Over Time', fontsize=16)
    plt.xlabel('Date and Time', fontsize=12)
    plt.ylabel(traffic_col, fontsize=12)

    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45)

    # Add grid lines for better readability
    plt.grid(True, linestyle='--', alpha=0.7)

    # Show the plot
    plt.tight_layout()
    plt.show()

    # Print some summary statistics
    print(f"\nSummary statistics of {traffic_col}:")
    print(df[traffic_col].describe())

    # Optional: Show hourly and daily patterns
    plt.figure(figsize=(15, 6))

    plt.subplot(1, 2, 1)
    sns.boxplot(x=df[datetime_col].dt.hour, y=df[traffic_col])
    plt.title('Hourly Pattern', fontsize=14)
    plt.xlabel('Hour of Day', fontsize=12)
    plt.ylabel(traffic_col, fontsize=12)

    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[datetime_col].dt.dayofweek, y=df[traffic_col])
    plt.title('Daily Pattern', fontsize=14)
    plt.xlabel('Day of Week (0=Monday, 6=Sunday)', fontsize=12)
    plt.ylabel(traffic_col, fontsize=12)

    plt.tight_layout()
    plt.show()

# Usage:
# Assuming 'df' is your DataFrame, 'datetime_column' is your datetime column,
# and 'TrafficIndexLive' is your traffic index column
visualize_traffic_over_time(df, 'Datetime', 'TrafficIndexLive')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure the Datetime column is in datetime format
df['Datetime'] = pd.to_datetime(df['Datetime'])

# Extract time and hour from Datetime
df['Time'] = df['Datetime'].dt.time
df['Hour'] = df['Datetime'].dt.hour

# Function to visualize TrafficIndexLive over time
def visualize_traffic_by_time(df):
    # Set up the plot
    plt.figure(figsize=(15, 8))

    # Create a scatter plot
    sns.scatterplot(data=df, x='Hour', y='TrafficIndexLive', alpha=0.5)

    # Add a trend line
    sns.regplot(data=df, x='Hour', y='TrafficIndexLive', scatter=False, color='red')

    # Customize the plot
    plt.title('TrafficIndexLive by Hour of Day', fontsize=16)
    plt.xlabel('Hour of Day', fontsize=12)
    plt.ylabel('TrafficIndexLive', fontsize=12)

    # Set x-axis to show all hours
    plt.xticks(range(0, 24))

    # Add grid lines for better readability
    plt.grid(True, linestyle='--', alpha=0.7)

    # Show the plot
    plt.tight_layout()
    plt.show()

    # Create a box plot to show distribution by hour
    plt.figure(figsize=(15, 8))
    sns.boxplot(x='Hour', y='TrafficIndexLive', data=df)
    plt.title('Distribution of TrafficIndexLive by Hour', fontsize=16)
    plt.xlabel('Hour of Day', fontsize=12)
    plt.ylabel('TrafficIndexLive', fontsize=12)
    plt.xticks(range(0, 24))
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

    # Print some summary statistics
    print("\nSummary statistics of TrafficIndexLive:")
    print(df['TrafficIndexLive'].describe())

    # Print average TrafficIndexLive by hour
    print("\nAverage TrafficIndexLive by Hour:")
    print(df.groupby('Hour')['TrafficIndexLive'].mean().sort_values(ascending=False))

# Call the function to visualize
visualize_traffic_by_time(df)
```

```python
df['Datetime'] = pd.to_datetime(df['Datetime'])

# Extract hour from Datetime
df['Hour'] = df['Datetime'].dt.hour

# Function to visualize average TrafficIndexLive by hour
def visualize_average_traffic_by_hour(df):
    # Calculate average TrafficIndexLive for each hour
    hourly_average = df.groupby('Hour')['TrafficIndexLive'].mean().reset_index()

    # Set up the plot
    plt.figure(figsize=(12, 6))

    # Create a line plot
    sns.lineplot(data=hourly_average, x='Hour', y='TrafficIndexLive', marker='o')

    # Customize the plot
    plt.title('Average TrafficIndexLive by Hour of Day', fontsize=16)
    plt.xlabel('Hour of Day', fontsize=12)
    plt.ylabel('Average TrafficIndexLive', fontsize=12)

    # Set x-axis to show all hours
    plt.xticks(range(0, 24))

    # Add grid lines for better readability
    plt.grid(True, linestyle='--', alpha=0.7)

    # Show the plot
    plt.tight_layout()
    plt.show()

    # Print average TrafficIndexLive by hour
    print("\nAverage TrafficIndexLive by Hour:")
    print(hourly_average.sort_values('Hour'))

# Call the function to visualize
visualize_average_traffic_by_hour(df)
```

```python
def visualize_average_traffic_by_date(df, start_date=None, end_date=None):
    if start_date:
        df = df[df['Date'] >= pd.to_datetime(start_date).date()]
    if end_date:
        df = df[df['Date'] <= pd.to_datetime(end_date).date()]

    daily_average = df.groupby('Date')['TrafficIndexLive'].mean().reset_index()

    plt.figure(figsize=(15, 6))
    sns.lineplot(data=daily_average, x='Date', y='TrafficIndexLive', marker='o')

    title = 'Average TrafficIndexLive by Date'
    if start_date and end_date:
        title += f' ({start_date} to {end_date})'
    plt.title(title, fontsize=16)

    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Average TrafficIndexLive', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

    print("\nAverage TrafficIndexLive by Date:")
    print(daily_average.sort_values('Date'))

    print("\nOverall Statistics:")
    print(daily_average['TrafficIndexLive'].describe())

# Example usage:
visualize_average_traffic_by_date(df)  # For all data
visualize_average_traffic_by_date(df, start_date='2023-07-01', end_date='2023-07-31')  # For a specific date range
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


def visualize_traffic_by_city(df):
    # Ensure City is treated as a category
    df['City'] = df['City'].astype('category')

    # 1. Box Plot
    plt.figure(figsize=(15, 8))
    sns.boxplot(x='City', y='TrafficIndexLive', data=df)
    plt.title('Distribution of TrafficIndexLive by City', fontsize=16)
    plt.xlabel('City', fontsize=12)
    plt.ylabel('TrafficIndexLive', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

    # 2. Bar Plot of Averages
    avg_traffic = df.groupby('City')['TrafficIndexLive'].mean().sort_values(ascending=False)

    plt.figure(figsize=(15, 8))
    avg_traffic.plot(kind='bar')
    plt.title('Average TrafficIndexLive by City', fontsize=16)
    plt.xlabel('City', fontsize=12)
    plt.ylabel('Average TrafficIndexLive', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

    # Print summary statistics
    print("\nSummary statistics of TrafficIndexLive for each City:")
    print(df.groupby('City')['TrafficIndexLive'].describe())

    # Perform one-way ANOVA
    from scipy import stats

    cities = df['City'].unique()
    samples = [df[df['City'] == city]['TrafficIndexLive'].values for city in cities]

    f_statistic, p_value = stats.f_oneway(*samples)
    print(f"\nOne-way ANOVA results:")
    print(f"F-statistic: {f_statistic}")
    print(f"p-value: {p_value}")

# Call the function
visualize_traffic_by_city(df)
```
