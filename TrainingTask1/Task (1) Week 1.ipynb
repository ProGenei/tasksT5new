```python
import seaborn as sns
from scipy.stats import chi2_contingency

def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))

categorical_features = df.select_dtypes(include=['object']).columns
categorical_features = [col for col in categorical_features if col != 'Time']

n = len(categorical_features)
cramer_matrix = np.zeros((n, n))

for i, feat1 in enumerate(categorical_features):
    for j, feat2 in enumerate(categorical_features):
        cramer_matrix[i, j] = cramers_v(df[feat1], df[feat2])

plt.figure(figsize=(12, 10))
sns.heatmap(cramer_matrix, annot=True, xticklabels=categorical_features, yticklabels=categorical_features, cmap='YlGnBu')
plt.title("Heatmap of Categorical Features (Cramer's V)")
plt.tight_layout()
plt.show()
```

```python
numerical_features = df.select_dtypes(include=['int64']).columns

for feature in numerical_features:
    # Get value counts and sort them
    value_counts = df[feature].value_counts().sort_index()

    # If there are too many unique values, group the least common ones
    if len(value_counts) > 10:
        top_n = value_counts.nlargest(9)
        others = pd.Series({'Others': value_counts.sum() - top_n.sum()})
        value_counts = top_n.append(others)

    # Create the pie chart
    plt.figure(figsize=(12, 8))
    plt.pie(value_counts.values, labels=value_counts.index, autopct='%1.1f%%', startangle=90)
    plt.title(f'Distribution of {feature}', fontsize=16)
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

    # Add a legend if there are many categories
    if len(value_counts) > 5:
        plt.legend(title=feature, loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

    plt.tight_layout()
    plt.show()

    # Print additional information
    print(f"\nUnique values in {feature}:", df[feature].nunique())
    print(f"Top 5 most common values in {feature}:")
    print(df[feature].value_counts().nlargest(5))
    print("\nSummary Statistics:")
    print(df[feature].describe())
    print("\n" + "="*50 + "\n")
```

### Explore relationships between variables (e.g., Age_band_of_driver vs. Accident_severity).


```python
def visualize_two_categories(df, cat1, cat2):
    # Create a contingency table
    cont_table = pd.crosstab(df[cat1], df[cat2])

    # Create the plot
    plt.figure(figsize=(12, 8))
    sns.heatmap(cont_table, annot=True, fmt='d', cmap='YlGnBu')

    # Customize the plot
    plt.title(f'Relationship between {cat1} and {cat2}', fontsize=16)
    plt.xlabel(cat2, fontsize=12)
    plt.ylabel(cat1, fontsize=12)

    # Show the plot
    plt.tight_layout()
    plt.show()

    # Print some additional information
    print(f"Total number of combinations: {cont_table.sum().sum()}")
    print(f"\nMost common combination:")
    most_common = cont_table.max().max()
    for i, row in cont_table.iterrows():
        for j, value in row.items():
            if value == most_common:
                print(f"{cat1}: {i}, {cat2}: {j}, Count: {value}")

# Usage:
# Assuming 'df' is your DataFrame, and 'category1' and 'category2' are your categorical columns
visualize_two_categories(df, 'Age_band_of_driver', 'Accident_severity')
```
